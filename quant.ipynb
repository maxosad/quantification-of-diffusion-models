```python
import torch
import torch.nn as nn
from diffusers import StableDiffusionPipeline, DDIMScheduler
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import time
import gc
import warnings
warnings.filterwarnings('ignore')

# Очистка памяти
torch.cuda.empty_cache()
gc.collect()

print("PyTorch версия:", torch.__version__)
print("CUDA доступен:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# ============================================================================
# 1. ЗАГРУЗКА И ПОДГОТОВКА МОДЕЛЕЙ
# ============================================================================

def load_model_fp32():
    """Загрузка модели в FP32"""
    print("Загрузка FP32 модели...")
    model = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float32,
        safety_checker=None,
        requires_safety_checker=False
    )
    return model

def load_model_fp16():
    """Загрузка модели в FP16"""
    print("Загрузка FP16 модели...")
    model = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16,
        safety_checker=None,
        requires_safety_checker=False
    )
    return model

def apply_dynamic_quantization_to_cpu(model):
    """Применение динамической квантации к модели на CPU"""
    print("Применение динамической INT8 квантации к модели...")
    
    # Перемещаем модель на CPU для квантации
    model = model.to("cpu")
    
    # Квантуем только линейные слои
    from torch.quantization import quantize_dynamic
    
    # Квантуем U-Net (только линейные слои)
    print("  Квантация U-Net...")
    model.unet = quantize_dynamic(
        model.unet,
        {torch.nn.Linear},  # Квантуем только линейные слои
        dtype=torch.qint8
    )
    
    # Квантуем Text Encoder (только линейные слои)
    print("  Квантация Text Encoder...")
    model.text_encoder = quantize_dynamic(
        model.text_encoder,
        {torch.nn.Linear},
        dtype=torch.qint8
    )
    
    # Квантуем VAE Encoder (только линейные слои)
    print("  Квантация VAE...")
    model.vae = quantize_dynamic(
        model.vae,
        {torch.nn.Linear},
        dtype=torch.qint8
    )
    
    return model

def apply_int8_quantization_with_torchao(model):
    """Применение INT8 квантации с помощью torchao (экспериментально)"""
    print("\nПрименение INT8 квантации с torchao...")
    try:
        import torchao
        
        # Квантуем модель на CPU
        model = model.to("cpu")
        
        # Преобразуем веса в INT8
        print("  Преобразование весов в INT8...")
        
        # Применяем квантацию к U-Net
        from torchao.quantization import apply_dynamic_quant
        model.unet = apply_dynamic_quant(model.unet)
        
        return model
    except ImportError:
        print("  torchao не установлен, используем стандартную квантацию")
        return apply_dynamic_quantization_to_cpu(model)

# ============================================================================
# 2. КЛАСС ДЛЯ СРАВНЕНИЯ МОДЕЛЕЙ
# ============================================================================

class DiffusionQuantizationBenchmark:
    def __init__(self):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.models = {}
        self.results = {}
        
    def setup_models(self):
        """Инициализация всех моделей"""
        print("="*60)
        print("ИНИЦИАЛИЗАЦИЯ МОДЕЛЕЙ")
        print("="*60)
        
        # FP32 модель
        print("\n1. FP32 модель:")
        fp32_model = load_model_fp32()
        if torch.cuda.is_available():
            fp32_model = fp32_model.to("cuda")
        self.models['fp32'] = fp32_model
        print(f"   Размещена на: {next(fp32_model.unet.parameters()).device}")
        
        # FP16 модель
        print("\n2. FP16 модель:")
        fp16_model = load_model_fp16()
        if torch.cuda.is_available():
            fp16_model = fp16_model.to("cuda")
        self.models['fp16'] = fp16_model
        print(f"   Размещена на: {next(fp16_model.unet.parameters()).device}")
        
        # INT8 модель (динамическая квантация)
        print("\n3. INT8 модель (CPU only):")
        int8_model = load_model_fp32()  # Загружаем как FP32
        int8_model = apply_dynamic_quantization_to_cpu(int8_model)
        # Оставляем на CPU - квантованные модели работают только на CPU!
        self.models['int8_cpu'] = int8_model
        print(f"   Размещена на: CPU (квантованные модели работают только на CPU)")
        
        # INT8 модель с ручной квантацией
        print("\n4. INT8 модель с ручной квантацией (эксперимент):")
        manual_int8_model = self.create_manual_int8_model()
        self.models['int8_manual'] = manual_int8_model
        
        return self.models
    
    def create_manual_int8_model(self):
        """Создание модели с ручной INT8 квантацией"""
        print("  Создание модели с ручной INT8 квантацией...")
        
        # Загружаем FP32 модель
        model = load_model_fp32()
        
        # Функция для квантации весов в INT8
        def quantize_weights_to_int8(module, layer_names=['weight', 'bias']):
            for name, param in module.named_parameters():
                if any(layer_name in name for layer_name in layer_names):
                    # Конвертируем в INT8 (симуляция)
                    if param.data is not None:
                        # Масштабируем и квантуем
                        scale = 127.0 / torch.max(torch.abs(param.data))
                        quantized = torch.clamp(param.data * scale, -128, 127).to(torch.int8)
                        param.data = quantized.float() / scale  # Возвращаем как float для совместимости
        
        # Применяем квантацию к основным компонентам
        print("  Квантация U-Net...")
        quantize_weights_to_int8(model.unet)
        
        print("  Квантация Text Encoder...")
        quantize_weights_to_int8(model.text_encoder)
        
        print("  Квантация VAE...")
        quantize_weights_to_int8(model.vae)
        
        return model
    
    def generate_with_model(self, model, prompt, seed=42, steps=20, device=None):
        """Генерация изображения с замером метрик"""
        if device is None:
            device = self.device
        
        # Устанавливаем устройство для генератора
        generator_device = "cpu" if "int8" in str(model.__class__).lower() else device
        generator = torch.Generator(device=generator_device).manual_seed(seed)
        
        # Замер времени
        start_time = time.time()
        
        try:
            # Для квантованных моделей на CPU
            if "int8" in str(model.__class__).lower() or next(model.unet.parameters()).device.type == "cpu":
                # Генерация на CPU
                with torch.no_grad():
                    image = model(
                        prompt,
                        num_inference_steps=steps,
                        generator=generator,
                        guidance_scale=7.5,
                        num_images_per_prompt=1
                    ).images[0]
            else:
                # Генерация на GPU с autocast для FP16
                if model.unet.dtype == torch.float16:
                    with torch.autocast(device_type="cuda"):
                        image = model(
                            prompt,
                            num_inference_steps=steps,
                            generator=generator,
                            guidance_scale=7.5,
                            num_images_per_prompt=1
                        ).images[0]
                else:
                    with torch.no_grad():
                        image = model(
                            prompt,
                            num_inference_steps=steps,
                            generator=generator,
                            guidance_scale=7.5,
                            num_images_per_prompt=1
                        ).images[0]
                        
        except Exception as e:
            print(f"  Ошибка генерации: {e}")
            # Создаем черное изображение как заглушку
            image = Image.new('RGB', (512, 512), color='black')
        
        gen_time = time.time() - start_time
        
        # Замер памяти
        if torch.cuda.is_available() and device == "cuda":
            torch.cuda.synchronize()
            memory_used = torch.cuda.max_memory_allocated() / 1024**2
        else:
            memory_used = 0
        
        return image, gen_time, memory_used
    
    def benchmark_models(self, prompts, steps=20):
        """Запуск бенчмарка для всех моделей"""
        print("\n" + "="*60)
        print("ЗАПУСК БЕНЧМАРКА")
        print("="*60)
        
        self.results = {}
        
        for model_name, model in self.models.items():
            print(f"\n{model_name.upper()}:")
            print("-" * 40)
            
            model_results = {
                'images': [],
                'times': [],
                'memory': [],
                'device': next(model.unet.parameters()).device.type
            }
            
            for i, prompt in enumerate(prompts):
                print(f"  Промпт {i+1}: {prompt[:50]}...")
                
                # Выбираем устройство для генерации
                device = "cpu" if "int8" in model_name else self.device
                
                image, gen_time, memory = self.generate_with_model(
                    model, prompt, steps=steps, device=device
                )
                
                model_results['images'].append(image)
                model_results['times'].append(gen_time)
                model_results['memory'].append(memory)
                
                print(f"    Время: {gen_time:.2f} сек, Память: {memory:.1f} MB")
            
            self.results[model_name] = model_results
        
        return self.results
    
    def visualize_results(self, prompts):
        """Визуализация результатов"""
        print("\n" + "="*60)
        print("ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ")
        print("="*60)
        
        n_models = len(self.models)
        n_prompts = len(prompts)
        
        fig, axes = plt.subplots(n_prompts, n_models, figsize=(4*n_models, 4*n_prompts))
        
        if n_prompts == 1:
            axes = axes.reshape(1, -1)
        
        model_names = list(self.models.keys())
        
        for i in range(n_prompts):
            for j, model_name in enumerate(model_names):
                ax = axes[i, j]
                
                if i < len(self.results[model_name]['images']):
                    ax.imshow(self.results[model_name]['images'][i])
                    
                    # Информация в заголовке
                    title = f"{model_name.upper()}\n"
                    title += f"Время: {self.results[model_name]['times'][i]:.2f}s\n"
                    title += f"Память: {self.results[model_name]['memory'][i]:.1f}MB\n"
                    title += f"Устройство: {self.results[model_name]['device']}"
                    
                    ax.set_title(title, fontsize=9)
                    ax.axis('off')
                else:
                    ax.text(0.5, 0.5, "No Image", ha='center', va='center')
                    ax.axis('off')
        
        plt.tight_layout()
        plt.show()
    
    def calculate_metrics(self):
        """Расчет метрик качества"""
        print("\n" + "="*60)
        print("РАСЧЕТ МЕТРИК КАЧЕСТВА")
        print("="*60)
        
        # Функция для расчета PSNR
        def calculate_psnr(img1, img2):
            if img1 is None or img2 is None:
                return 0
            img1_np = np.array(img1).astype(np.float32)
            img2_np = np.array(img2).astype(np.float32)
            
            mse = np.mean((img1_np - img2_np) ** 2)
            if mse == 0:
                return float('inf')
            max_pixel = 255.0
            return 20 * np.log10(max_pixel / np.sqrt(mse))
        
        # Используем FP32 как референс
        reference_images = self.results.get('fp32', {}).get('images', [])
        
        if not reference_images:
            print("Нет референсных изображений FP32")
            return
        
        print("\nСравнение с FP32 (PSNR в dB):")
        print("-" * 40)
        
        for model_name in self.models.keys():
            if model_name == 'fp32':
                continue
                
            model_images = self.results.get(model_name, {}).get('images', [])
            
            if model_images:
                psnr_values = []
                for i, (ref_img, test_img) in enumerate(zip(reference_images, model_images)):
                    if i < len(model_images):
                        psnr = calculate_psnr(ref_img, test_img)
                        psnr_values.append(psnr)
                
                if psnr_values:
                    avg_psnr = np.mean(psnr_values)
                    print(f"{model_name.upper()}: средний PSNR = {avg_psnr:.2f} dB")
    
    def print_summary(self):
        """Вывод сводной информации"""
        print("\n" + "="*60)
        print("СВОДКА РЕЗУЛЬТАТОВ")
        print("="*60)
        
        # Собираем статистику
        summary_data = []
        
        for model_name in self.models.keys():
            times = self.results.get(model_name, {}).get('times', [])
            memory = self.results.get(model_name, {}).get('memory', [])
            device = self.results.get(model_name, {}).get('device', 'unknown')
            
            if times:
                avg_time = np.mean(times)
                avg_memory = np.mean(memory) if memory else 0
                
                summary_data.append({
                    'model': model_name,
                    'avg_time': avg_time,
                    'avg_memory': avg_memory,
                    'device': device
                })
        
        # Вывод таблицы
        print("\nМодель           | Устройство | Время (сек) | Память (MB)")
        print("-" * 60)
        
        for data in summary_data:
            print(f"{data['model']:15} | {data['device']:10} | {data['avg_time']:10.2f} | {data['avg_memory']:10.1f}")
        
        # Сравнение относительно FP32
        fp32_data = next((d for d in summary_data if d['model'] == 'fp32'), None)
        
        if fp32_data:
            print("\nСравнение относительно FP32:")
            print("-" * 40)
            
            for data in summary_data:
                if data['model'] != 'fp32':
                    time_ratio = fp32_data['avg_time'] / data['avg_time'] if data['avg_time'] > 0 else 0
                    mem_ratio = fp32_data['avg_memory'] / data['avg_memory'] if data['avg_memory'] > 0 else 0
                    
                    print(f"{data['model']}: ускорение = {time_ratio:.2f}x, экономия памяти = {mem_ratio:.2f}x")

# ============================================================================
# 3. ЗАПУСК ЭКСПЕРИМЕНТА
# ============================================================================

def main():
    """Основная функция эксперимента"""
    
    # Инициализация бенчмарка
    benchmark = DiffusionQuantizationBenchmark()
    
    # Настройка моделей
    models = benchmark.setup_models()
    
    # Тестовые промпты
    prompts = [
        "A beautiful sunset over mountains",
        "A cute cat sitting on a windowsill",
        "An astronaut riding a horse on Mars"
    ]
    
    # Запуск бенчмарка
    results = benchmark.benchmark_models(prompts, steps=15)
    
    # Визуализация
    benchmark.visualize_results(prompts)
    
    # Расчет метрик
    benchmark.calculate_metrics()
    
    # Сводка
    benchmark.print_summary()
    
    # Дополнительный анализ
    print("\n" + "="*60)
    print("РЕКОМЕНДАЦИИ ПО ИСПОЛЬЗОВАНИЮ КВАНТАЦИИ")
    print("="*60)
    
    print("""
1. FP32 (полная точность):
   - Лучшее качество
   - Максимальное использование памяти
   - Использовать для финального рендеринга

2. FP16 (половинная точность):
   - Качество почти как у FP32
   - 2x экономия памяти
   - 1.5-2x ускорение на GPU
   - Рекомендуется для большинства задач на GPU

3. INT8 квантация:
   - Работает ТОЛЬКО на CPU
   - 4x экономия памяти
   - 2-3x ускорение на CPU
   - Заметная потеря качества
   - Использовать:
     * Когда GPU недоступен
     * Для мобильных устройств
     * При ограничениях по памяти

4. Важные ограничения:
   - PyTorch INT8 квантация работает только на CPU
   - Для GPU нужны специальные ядра (TensorRT, cuDNN)
   - Качество INT8 сильно зависит от калибровки
    """)
    
    # Очистка памяти
    print("\nОчистка памяти...")
    for model_name, model in models.items():
        del model
    torch.cuda.empty_cache()
    gc.collect()
    
    print("\nЭксперимент завершен!")

# ============================================================================
# 4. АЛЬТЕРНАТИВНЫЙ ПОДХОД: КВАНТАЦИЯ ЧЕРЕЗ ONNX
# ============================================================================

def run_onnx_quantization_experiment():
    """Эксперимент с ONNX квантацией (работает на CPU)"""
    print("\n" + "="*60)
    print("ЭКСПЕРИМЕНТ С ONNX КВАНТАЦИЕЙ")
    print("="*60)
    
    try:
        from optimum.onnxruntime import ORTStableDiffusionPipeline
        import onnxruntime as ort
        
        print("1. Экспорт модели в ONNX...")
        
        # Создание пайплайна ONNX
        model_id = "runwayml/stable-diffusion-v1-5"
        
        # Загрузка ONNX модели (если уже экспортирована)
        print("2. Загрузка ONNX модели...")
        
        # Пробуем загрузить существующую ONNX модель
        try:
            onnx_pipeline = ORTStableDiffusionPipeline.from_pretrained(
                model_id,
                provider="CPUExecutionProvider"
            )
            
            print("3. Генерация тестового изображения...")
            prompt = "A beautiful sunset"
            image = onnx_pipeline(prompt).images[0]
            
            # Сохранение
            image.save("onnx_generated_image.png")
            print(f"   Изображение сохранено: onnx_generated_image.png")
            
            # Размер модели
            import os
            model_size = os.path.getsize("onnx_generated_image.png") / 1024
            print(f"   Размер сгенерированного изображения: {model_size:.1f} KB")
            
        except Exception as e:
            print(f"   Ошибка: {e}")
            print("   Пропускаем ONNX эксперимент")
            
    except ImportError:
        print("ONNX Runtime не установлен. Пропускаем эксперимент.")

# ============================================================================
# ЗАПУСК ЭКСПЕРИМЕНТА
# ============================================================================

if __name__ == "__main__":
    print("НАЧАЛО ЭКСПЕРИМЕНТА ПО КВАНТАЦИИ ДИФФУЗИОННЫХ МОДЕЛЕЙ")
    print("="*60)
    
    # Основной эксперимент
    main()
    
    # Дополнительный эксперимент с ONNX
    run_onnx_quantization_experiment()
    
    print("\n" + "="*60)
    print("ВЫВОДЫ:")
    print("="*60)
    print("""
    1. INT8 квантация в PyTorch работает ТОЛЬКО на CPU
    2. Для GPU нужно использовать другие библиотеки (TensorRT, etc.)
    3. FP16 дает лучший баланс скорости и качества на GPU
    4. INT8 полезен для развертывания на CPU и мобильных устройствах
    5. Качество INT8 обычно ниже, чем у FP16/FP32
    """)
```
